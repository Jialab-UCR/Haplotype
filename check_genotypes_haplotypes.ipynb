```python
#############################
#### samtools

import os
import glob
from fnmatch import fnmatch
from collections import OrderedDict
import re

hetSNP = {}
os.chdir('/home/ruidong/Documents/Research/Haplotype/Data/Version2/GenomeResearch/samtools/')
```

```python
snps = OrderedDict()
with open('Phased_haplotypes.chr1.txt') as f:
    for line in f:
        line = line.rstrip()
        lst = line.split('\t')
        pos = lst[0]+':'+lst[1]
        snps[pos] = ['NA']*11

[name for name in os.listdir('.') if fnmatch(name, '*.samtools.snp.raw.chr1.vcf')]
vcfFiles = sorted(glob.glob('*.samtools.snp.raw.chr1.vcf'))
#vcfNames = [file.split('.')[0] for file in vcfFiles]

samples = []

i = -1
for fl in vcfFiles:
    i += 1
    with open(fl) as f:
        for line in f:
            line = line.rstrip()
            lst = line.split('\t')
            if lst[0] == '#CHROM':
                samples.append(lst[-1])
                
            elif lst[0] == '1':
                pos = lst[0]+':'+lst[1]
                genotype = lst[-1].split(':')[0]
                
                if pos not in snps.keys():
                    continue
                
                if float(lst[5]) <= 20:
                    continue
                
                if re.search('DP=(\d+)', lst[7]):
                    mch = re.search('DP=(\d+)', lst[7])
                    dp = int(mch.group(1))
                    if dp < 3:
                        continue
                    
                if genotype == '0/0':
                    snps[pos][i] = '0'
                elif genotype == '1/1':
                    snps[pos][i] = '1'
                
                    
fh = open('sperm11.samtools.chr1.Q20.DP2.filter.txt', 'wt')            
fh.write('\t'+'\t'.join(samples)+'\n')
                    
for ke, val in snps.items():
    fh.write(ke+'\t'+'\t'.join(val)+'\n')
    
fh.close()
```

```python
########################
### check crossover

fh = open('crossover.count.txt', 'wt')
cv = OrderedDict()
with open('cvDa.txt') as f:
    for line in f:
        line = line.rstrip()
        lst = line.split('\t')
        
        if lst[0] == '':
            fh.write('\tCounts\t'+line+'\n')
            continue
        else:
            for i in range(len(lst)):
                if lst[i] not in cv.keys():
                    cv[lst[i]] = ['']*11
                    cv[lst[i]][i] = 'Y'
                else:
                    cv[lst[i]][i] = 'Y'
                    
for ke, val in cv.items():
    fh.write('\t'.join([ke,str(len([y for y in val if y == 'Y']))]+val)+'\n')
    
fh.close()
```
